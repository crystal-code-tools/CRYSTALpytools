# Properties output and plot

Band plot

```python
from CRYSTALpytools.crystal_io import Properties_output
from CRYSTALpytools.plot import plot_cry_band, plot_electron_band

obj1 = Properties_output().read_cry_band('electronic_structure/band_hTaAs.BAND')
obj2 = Properties_output().read_cry_band('electronic_structure/band_hTaAs_noso.BAND')
obj3 = Properties_output().read_cry_band('electronic_structure/band_hTaAs_P15.BAND')
obj4 = Properties_output().read_cry_band('electronic_structure/band_hTaAs_P25.BAND')

fig = plot_cry_band([obj1, obj2, obj3, obj4], energy_range=[-5, 5],
                    k_labels=['Gamma', 'M', 'K', 'A', 'L', 'H', 'K'],
                    mode='multi', labels=['NOSOC', 'SCDFT', 'P15', 'P25'],
                    color=['b', 'c', 'g', 'r'],
                    linestl=['-','--','dotted','dashdot'])

fig = plot_electron_band([obj1, obj2, obj3, obj4], energy_range=[-5, 5],
                         k_labels=['Gamma', 'M', 'K', 'A', 'L', 'H', 'K'],
                         mode='compare', labels=['NOSOC', 'SCDFT', 'P15', 'P25'])
```

Density of states

```python
from CRYSTALpytools.crystal_io import Properties_output
from CRYSTALpytools.plot import plot_cry_doss, plot_electron_dos

obj1 = Properties_output().read_cry_doss('electronic_structure/doss_96.DOSS')
obj2 = Properties_output().read_cry_doss('electronic_structure/doss_cu.DOSS')

fig = plot_cry_doss(obj1, linestl=['-','--','dotted'], fermi='indigo', color='orange',
                    prj=[1,2,3], energy_range=[-5, 5], dos_range=[0, 10])
fig = plot_electron_dos(obj2, energy_range=[-5, 5], dos_range=[-10, 10], fermi='indigo',
                        overlap=True, beta='down', linewidth=1.5,
                        color=['blue', 'orange', 'limegreen', 'purple'],
                        labels=['prj1', 'prj2', 'prj4', 'Tot'], prj=[1,2,4,5])
```
